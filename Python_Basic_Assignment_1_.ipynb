Q 3. Compare and contrast mutable and immutable objects in Python with examples.

  
 Mutable    vs  Immutable in Python
  
   
      Feature	                    Mutable	                       Immutable

     Can be changed?	     Yes, after creation	         No, cannot be changed once created
     Memory address (ID)	Remains the same when modified	Changes if "modified" (new object)

     Examples	          list, dict, set, bytearray	          int, float, str, tuple, bool, frozenset


 Mutable Objects

Mutable objects can be modified in-place without creating a new object.

fruits = ['apple', 'banana']

print(id(fruits))  # Memory ID before modification

fruits.append('cherry')

print(fruits)      # ['apple', 'banana', 'cherry']

print(id(fruits))  # Same ID - object modified in-place


Immutable Objects

Immutable objects cannot be changed once created. Any "modification" actually creates a new object.

 Example: str

 name = "Alice"

print(id(name))  # Original ID

name += " Smith"

print(name)      # 'Alice Smith'

print(id(name))  # New ID - new string created

Example: int

x = 10

print(id(x))

x += 1

print(x)         # 11

print(id(x))     # New ID - x now refers to a new int object

 Q 5. Explain the concept of type casting in Python with examples.

  Type casting means converting a value from one data type to another.
  In Python, this is often called type conversion.

  There are two types:

a. Explicit Type Casting (manual)

  the programmer, deliberately convert one type to another using built-in functions like int(), float(), str(), etc.

b. Implicit Type Casting (automatic)

 Python automatically converts types when needed (e.g., during calculations).

  Explicit Type Casting Examples:

- Convert str to int

  num_str = "10"

  num_int = int(num_str)

  print(num_int + 5)  # Output: 15

  -Convert float to int

  price = 99.99

  whole_price = int(price)

  print(whole_price)  # Output: 99
      
-Convert int to float

  x = 7
       
  y = float(x)

  print(y)  # Output: 7.0

-Convert int to str

  age = 25

  age_str = str(age)

  print("Age is " + age_str)  # Output: Age is 25
        
-Convert list to tuple

 my_list = [1, 2, 3]

 my_tuple = tuple(my_list)

 print(my_tuple)  # Output: (1, 2, 3)

Implecit type casting exemple

 x = 10      # int

 y = 2.5     # float

 print(result)       # Output: 12.5

 print(type(result)) # Output: <class 'float'>
